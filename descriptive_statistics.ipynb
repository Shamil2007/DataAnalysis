# Descriptive Statistics Exercises and Solutions
This notebook includes basic exercises related to descriptive statistics, including mean, median, mode, variance, and standard deviation.

## Exercise 1: Calculate Measures of Central Tendency
**Task:** Given the following data, calculate the mean, median, and mode:
`data = [12, 15, 12, 18, 19, 12, 16, 14, 18, 19]`

```python
import statistics as stats
import numpy as np
import pandas as pd

data = [12, 15, 12, 18, 19, 12, 16, 14, 18, 19]
mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data)
mean, median, mode
```

## Exercise 2: Calculate Measures of Dispersion
**Task:** Using the same data, calculate the variance and standard deviation.

```python
variance = np.var(data, ddof=1)  # Sample variance
std_dev = np.std(data, ddof=1)  # Sample standard deviation
variance, std_dev
```

**Task:** Compute the minimum, Q1, median (Q2), Q3, and maximum.

```python
min_val = np.min(data)
q1 = np.percentile(data, 25)
q2 = np.percentile(data, 50)
q3 = np.percentile(data, 75)
max_val = np.max(data)
min_val, q1, q2, q3, max_val
```

In a small company, the monthly salaries (in hundreds) are: `[35, 40, 45, 50, 40, 35, 100, 45, 40, 35]`. Compute the **mean**, **median**, and **mode** and comment on any skewness.

```python
def find_mean(arr):
    total = 0
    for num in arr:
        total += num
    return total / len(arr)
```

```python
def find_median(arr):
    arr.sort()

    length = len(arr)
    mid = length // 2
    if length % 2 == 0:
        return (arr[mid] + arr[mid - 1]) / 2
    return arr[mid]
```

```python
def find_mode(arr):
    num_freq = {}

    for num in arr:
        num_freq[num] = num_freq.get(num, 0) + 1

    max_count = max(num_freq.values())
    return [num for num, count in num_freq.items() if count == max_count]
```

In other words, a few exceptionally high salaries are inflating the mean, creating the impression of higher typical pay than what most employees actually receive.
This is important for analysis because decisions based only on the mean (like setting salary benchmarks) could overestimate what‚Äôs ‚Äúnormal‚Äù in the company.

## üìå Homework: Measures of Dispersion
#### Task 1:
Use the dataset: `[5, 10, 10, 10, 15, 20, 25, 25, 30]` to calculate **variance** and **standard deviation**.

#### Task 2:
Income levels (in AZN) for 8 families are: `[500, 700, 800, 600, 1200, 1000, 950, 1100]`. Calculate **variance** and **standard deviation**.

#### Task 3:
Evaluate the spread of scores in a classroom: `[60, 65, 70, 75, 80, 85, 90, 95, 100]`. What are the **sample variance** and **standard deviation**?

####The variance formula:

image.png

```python
def find_variance(arr):
    total = 0
    mean = find_mean(arr)
    for num in arr:
        difference = num - mean
        total += difference ** 2
    return total / (len(arr) - 1)
```

####Standard Deviation

image.png

```python
def find_std(arr):
    return np.sqrt((find_variance(arr)))
```

###Task 1

```python
arr = [5, 10, 10, 10, 15, 20, 25, 25, 30]

f"Variance {np.var(arr)}, Standard Deviation {np.std(arr)}"
```

```python
f"Variance {find_variance(arr)}, Standard Deviation {find_std(arr)}"
```

The difference you‚Äôre seeing comes from how the variance and standard deviation are calculated:

- My own function uses the sample formula, which divides by (n ‚àí 1).

- The default library function calculate for the entire population, dividing by n.

Because dividing by (n ‚àí 1) makes the denominator slightly smaller, the sample variance/standard deviation is usually a bit larger than the population version.
Understanding this distinction is key when deciding whether you‚Äôre analyzing a sample of a bigger group or the whole population.

###Task 3

```python
scores = [60, 65, 70, 75, 80, 85, 90, 95, 100]

f"Variance {find_variance(scores)}, Standard Deviation {find_std(scores)}"
```

```python
f"Mean: {find_mean(scores)}"
```

- Mean (80): On average, the scores cluster around 80.

- Variance (187.5): Variance measures how spread out the scores are from the mean.

    - A variance of 187.5 means that, on average, each score‚Äôs squared distance from 80 is about 187.5.

    - Because it‚Äôs in squared units, it‚Äôs not as intuitive by itself.

- Standard Deviation (‚âà13.69): The standard deviation is the square root of variance, bringing it back to the same unit as the scores.

    - Here, a standard deviation of about 13.7 means most scores fall roughly 13‚Äì14 points above or below the mean.

    - In a roughly normal distribution, about 68% of scores would be between 66 and 94 (80 ¬± 13.7).

## üìå Homework: Five Number Summary
#### Task 1:
Compute the **five-number summary** for `[100, 102, 105, 107, 110, 113, 115, 117, 120]`.

#### Task 2:
Given household expenses in a community: `[230, 250, 270, 290, 310, 330, 350, 370, 390, 410]`, calculate the **minimum**, **Q1**, **Q2**, **Q3**, and **maximum**.

#### Task 3:
Analyze the five-number summary for the number of working hours per week: `[35, 36, 38, 40, 42, 44, 46, 48, 50, 55]`.

```python
def summary(arr):
    arr.sort()

    print(f"Min: {np.min(arr)}, Max: {np.max(arr)}")
    print(f"Median: {find_median(arr)}")
    print(f"Q1: {np.percentile(arr, 25)}")
    print(f"Q3: {np.percentile(arr, 75)}")
    return
```
