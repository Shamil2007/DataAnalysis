```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#Linear Plot

```python
# Median Developer Salaries by Age
ages_x = [18, 23, 27, 30, 33, 39, 42, 43, 45, 49, 55]

dev_y = [38496, 42000, 46752, 49320, 53200,
         56000, 62316, 64928, 67317, 68748, 73752]


plt.plot(ages_x, dev_y) # draws a line plot, connecting data points (ages_x vs dev_y).
plt.show() # displays the graph window.
```

##Labels

```python
plt.plot(ages_x, dev_y)
plt.xlabel("Ages") # adds a label to the x-axis
plt.ylabel("Median Salary (USD)") # adds a label to the y-axis
plt.title("Median Salary (USD) by Age") # adds a title above the graph

plt.show()
```

##ADD Data

```python
# Median Python Developer Salaries by Age
py_dev_y = [45372, 48876, 53850, 57287, 63016,
            65998, 70003, 70000, 71496, 75370, 83640]

plt.plot(ages_x, py_dev_y)

# Median Developers
plt.plot(ages_x, dev_y)

plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.title("Median Salary (USD) by Age")
plt.show()
```

Great, but we can not identify which python or median developer salary

##Lengend

# Pie Charts

```python
slices = [30, 70]

plt.pie(slices) # draws a pie chart.

plt.title("Pie chart Intro")
plt.tight_layout()
plt.show()
```

##Label

```python
labels = ["Developers", "Others"]
slices = [30, 70] # Each value in slices represents a portion of the circle.

plt.pie(slices, labels=labels) # assigns category names to each slice.

plt.title("Pie chart Intro")
plt.tight_layout()
plt.show()
```

```python
labels = ["Developers", "Others"]

#The numbers sum sholdn't be always equal to 100
slices = [30, 50]

plt.pie(slices, labels=labels)

plt.title("Pie chart Intro")
plt.tight_layout()
plt.show()
```

## Wedge color and size

```python
labels = ["Developers", "Others"]
slices = [30, 70]

plt.pie(slices, labels=labels, wedgeprops={'edgecolor': 'Green', 'linewidth': 2})
# edgecolor -> controls the appearance of pie slices
# edgecolor='Green' → sets border color of slices.
# linewidth=3 → makes borders thicker.

plt.title("Pie chart Intro")
plt.tight_layout()
plt.show()
```

## More slices

```python
labels = ["Developers", "Marketings", "Accountants", "Salers"]
slices = [30, 20, 10, 50]

plt.pie(slices, labels=labels, wedgeprops={'edgecolor': 'Black', 'linewidth': 2})

plt.title("Pie chart Intro")
plt.tight_layout()
plt.show()
```

##Colors

```python
colors = ['r', 'b', 'y', 'lightblue']

plt.pie(slices, labels=labels, wedgeprops={'edgecolor': 'Black', 'linewidth': 2}, colors=colors)

plt.title("Pie chart Intro")
plt.tight_layout()
plt.show()
```

##Explode

```python
labels = ["Developers", "Marketings", "Accountants", "Salers"]
explodes = [0, 0, 0.1, 0] # separates slices from the pie

plt.pie(slices, labels=labels, explode=explodes,
        wedgeprops={'edgecolor': 'black', 'linewidth': 1.5})

plt.title("Pie chart Intro")
plt.tight_layout()
plt.show()
```

##Percentage

```python
plt.pie(slices, labels=labels,
        autopct='%1.2f%%',
        wedgeprops={'edgecolor': 'black', 'linewidth': 1.5})

# autopct='%1.2f%%' → shows percentage values on slices.

# %1.2f%% → display with 2 decimal places (e.g., 25.00%).

# %1.0f%% → no decimals (e.g., 25%).

plt.title("Pie chart Intro")
plt.tight_layout()
plt.show()
```

#Histograms

##Bins

```python
ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]

plt.hist(ages, bins=20)
# bins=20 → splits the data into 20 intervals.

# More bins → more detailed view.

# Fewer bins → more general view

plt.title('Ages of Respondents')
plt.xlabel('Ages')
plt.ylabel('Total Respondents')

plt.tight_layout()

plt.show()
```

## Logarithmic scale


```python
df = pd.read_csv("data.csv")

ages = df["Age"]

plt.hist(ages, bins=20, edgecolor='black', log=True)
# edgecolor='black' → adds borders to bars for clarity.

# log=True → y-axis uses logarithmic scale instead of linear.

# Useful when values vary widely (to avoid tall bars hiding small ones

plt.title('Ages of Respondents')
plt.xlabel('Ages')
plt.ylabel('Total Respondents')

plt.tight_layout()

plt.show()
```

##Vertical Line

```python
df = pd.read_csv("data.csv")

ages = df["Age"]

plt.hist(ages, bins=20, edgecolor='black', log=True)

plt.title('Ages of Respondents')
plt.xlabel('Ages')
plt.ylabel('Total Respondents')

median_age = ages.median()

plt.axvline(median_age, color="black")
# plt.axvline() → draws a vertical reference line.

# Here, it marks the median age.
# color="black" → sets the line color.

plt.tight_layout()

plt.show()
```

#SubPlots

```python
plt.subplot(2, 2, 1)  # (rows, cols, position)
plt.plot([1, 2, 3], [1, 4, 9], 'r')   # First subplot

plt.subplot(2, 2, 2)
plt.plot([1, 2, 3], [1, 2, 3], 'g')   # Second subplot

plt.subplot(2, 2, 3)
plt.plot([1, 2, 3], [1, 8, 27], 'b')  # Third subplot

plt.subplot(2, 2, 4)
plt.plot([1, 2, 3], [3, 2, 1], 'k')   # Fourth subplot

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(2, 2)  # 2x2 grid

ax[0, 0].plot([1, 2, 3], [1, 4, 9], 'r')
ax[0, 1].plot([1, 2, 3], [1, 2, 3], 'g')
ax[1, 0].plot([1, 2, 3], [1, 8, 27], 'b')
ax[1, 1].plot([1, 2, 3], [3, 2, 1], 'k')

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, sharex=True, sharey=True)

# sharex=True → all subplots share the same x-axis.

# sharey=True → all subplots share the same y-axis.

ax[0, 0].hist([1, 2, 2, 3])
ax[0, 1].hist([2, 3, 3, 4])
ax[1, 0].hist([3, 4, 4, 5])
ax[1, 1].hist([4, 5, 5, 6])

plt.show()
```
